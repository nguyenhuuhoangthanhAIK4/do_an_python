```python
import pygame
import random

# Khởi tạo pygame
pygame.init()

# Thiết lập màn hình game
width, height = 640, 480
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Snake Game")

# Các màu sắc
black = (0, 0, 0)
white = (255, 255, 255)
green = (0, 255, 0)
red = (255, 0, 0)

# Các biến cho trò chơi
block_size = 20
snake_speed = 15

# Hàm vẽ rắn
def draw_snake(snake_body):
    for block in snake_body:
        pygame.draw.rect(screen, green, [block[0], block[1], block_size, block_size])

# Hàm chạy trò chơi
def game_loop():
    game_over = False
    game_close = False

    # Vị trí ban đầu của con rắn
    snake_x = width // 2
    snake_y = height // 2

    # Độ dài ban đầu của con rắn và tốc độ di chuyển
    snake_body = [[snake_x, snake_y]]
    length_of_snake = 1

    # Vị trí ban đầu của thức ăn
    food_x = round(random.randrange(0, width - block_size) / block_size) * block_size
    food_y = round(random.randrange(0, height - block_size) / block_size) * block_size

    # Điều khiển ban đầu
    x_change = 0
    y_change = 0

    # Vòng lặp chính
    while not game_over:
        while game_close == True:
            # Hiển thị điểm số khi game over
            screen.fill(black)
            font = pygame.font.SysFont('comicsansms', 35)
            text = font.render('Game Over! Press Q-Quit or C-Play Again', True, white)
            text_rect = text.get_rect(center=(width / 2, height / 2))
            screen.blit(text, text_rect)

            pygame.display.update()

            # Xử lý sự kiện khi game over
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    elif event.key == pygame.K_c:
                        game_loop()

        # Xử lý sự kiện
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x_change = -block_size
                    y_change = 0
                elif event.key == pygame.K_RIGHT:
                    x_change = block_size
                    y_change = 0
                elif event.key == pygame.K_UP:
                    x_change = 0
                    y_change = -block_size
                elif event.key == pygame.K_DOWN:
                    x_change = 0
                    y_change = block_size

        # Điều chỉnh vị trí của rắn
        snake_x += x_change
        snake_y += y_change

        # Xử lý va chạm với ranh giới màn hình
        if snake_x >= width or snake_x < 0 or snake_y >= height or snake_y < 0:
            game_close = True
        # Xử lý va chạm với thân rắn
        for segment in snake_body[1:]:
            if segment == [snake_x, snake_y]:
                game_close = True

        # Thêm đầu của rắn vào đầu danh sách
        snake_head = [snake_x, snake_y]
        snake_body.insert(0, snake_head)

        # Xử lý khi ăn thức ăn
        if snake_x == food_x and snake_y == food_y:
            length_of_snake += 1
            food_x = round(random.randrange(0, width - block_size) / block_size) * block_size
            food_y = round(random.randrange(0, height - block_size) / block_size) * block_size
        else:
            snake_body.pop()

        # Hiển thị các đối tượng trong game
        screen.fill(black)
        pygame.draw.rect(screen, red, [food_x, food_y, block_size, block_size])
        draw_snake(snake_body)

        pygame.display.update()

        # Tốc độ của rắn
        pygame.time.Clock().tick(snake_speed)

    pygame.quit()
    quit()

# Khởi động game
game_loop()
```
